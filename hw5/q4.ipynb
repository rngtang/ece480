```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf_Y(y):
    return (1 / np.sqrt(2 * np.pi * y)) * np.exp(-y / 2)
```

```python
# Generate a range of y values for plotting the analytic PDF
y_values = np.linspace(0.01, 10, 1000)  # Start from a small positive number to avoid division by zero
pdf_values = pdf_Y(y_values)
```

```python
# Step (i): Draw 1000 observations of the random variable X ~ N(0, 1)
np.random.seed(42)  # For reproducibility
X_samples = np.random.normal(0, 1, 1000)
```

```python
# Step (ii): Transform the observations to the RV Y = X^2
Y_samples = X_samples ** 2
```

```python
# Plot the histogram of the simulated Y values (normalized to integrate to 1)
plt.figure(figsize=(8, 6))
plt.hist(Y_samples, bins=50, density=True, alpha=0.6, label="Simulated Y = X^2", color='orange')

# Plot the analytic PDF of Y for comparison
plt.plot(y_values, pdf_values, label='Analytic PDF of Y', color='blue')

# Title and labels
plt.title('Comparison of Simulated and Analytic PDF of Y = X^2 where X ~ N(0, 1)')
plt.xlabel('y')
plt.ylabel('Density')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
